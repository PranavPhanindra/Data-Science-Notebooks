```python
import matplotlib.pyplot as plt
import random
from matplotlib import style
%matplotlib inline
import pandas as pd
```

```python
web_customers = random.sample(range(900,2000),20)
print(web_customers)
```

```python
time_hrs = random.sample(range(1,24),20)
time_hrs
```

```python
Web_Customers = pd.DataFrame({'Time(hrs)' : time_hrs , 'Customer_Count' : web_customers},index = time_hrs)
Web_Customers = Web_Customers.sort_index()
Web_Customers
```

```python
style.available
style.use('ggplot')
plt.plot(Web_Customers['Time(hrs)'],Web_Customers['Customer_Count'],linestyle = '-',color='b',linewidth = 1)
plt.title('Web Traffic')
plt.xlabel('Hrs')
plt.ylabel('User Count')
plt.show()
```

```python

style.use('ggplot')
plt.plot(Web_Customers['Time(hrs)'],Web_Customers['Customer_Count'],linestyle = '-',color='b',linewidth = 1)
plt.scatter(time_hrs,web_customers)
plt.axis([5,20,50,2000])
plt.title('Web Traffic')
plt.xlabel('Hrs')
plt.ylabel('User Count')
plt.legend()

plt.show()
```

```python
Day_1 = random.sample(range(900,2000),10)
Day_2 = random.sample(range(900,2000),10)
Day_3 = random.sample(range(900,2000),10)
Day_4 = random.sample(range(900,2000),10)
time_hrs = random.sample(range(1,24),10)

Timeline_DataFrame = pd.DataFrame({'Time':time_hrs,'Day_1':Day_1,'Day_2':Day_2 , 'Day_3':Day_3 , 'Day_4':Day_4}, index = time_hrs)
Timeline_DataFrame = Timeline_DataFrame.sort_index()
Timeline_DataFrame
```

```python
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=0.25)
plt.subplot(1,2,1)
plt.title('Day-1')
plt.plot(Timeline_DataFrame['Time'],Timeline_DataFrame['Day_1'],linestyle = '-',color='b',linewidth = 1,alpha =0.4)
plt.xlabel('Hrs')
plt.subplot(1,2,2)
plt.title('Day-2')
plt.plot(Timeline_DataFrame['Time'],Timeline_DataFrame['Day_2'],linestyle = '-',color='b',linewidth = 1,alpha =0.4)
plt.xlabel('Hrs')
```

```python
plt.figure(figsize=(6,6))
plt.subplots_adjust(wspace=1)
plt.subplot(2,1,1)
plt.title('Day-1')
plt.plot(Timeline_DataFrame['Time'],Timeline_DataFrame['Day_1'],linestyle = '-',color='b',linewidth = 1,alpha =0.4)
plt.xlabel('Hrs')
plt.subplot(2,1,2)
plt.title('Day-2')
plt.plot(Timeline_DataFrame['Time'],Timeline_DataFrame['Day_2'],linestyle = '-',color='b',linewidth = 1,alpha =0.4)
plt.xlabel('Hrs')
```

```python
style.use('ggplot')
plt.figure(figsize = (9,9))
plt.subplots_adjust(hspace=0.3)
plt.subplot(2,2,1)
plt.title('Day-1') 
plt.plot(Timeline_DataFrame['Time'],Timeline_DataFrame['Day_1'],linestyle = '-',color='b',linewidth = 1,alpha =0.4)
plt.subplot(2,2,2)
plt.title('Day-2') 
plt.plot(Timeline_DataFrame['Time'],Timeline_DataFrame['Day_2'],linestyle = '-',color='g',linewidth = 1,alpha =0.4)
plt.subplot(2,2,3)
plt.title('Day-3') 
plt.plot(Timeline_DataFrame['Time'],Timeline_DataFrame['Day_3'],linestyle = '-',color='m',linewidth = 1,alpha =0.4)
plt.subplot(2,2,4)
plt.title('Day-4') 
plt.plot(Timeline_DataFrame['Time'],Timeline_DataFrame['Day_4'],linestyle = '-',color='r',linewidth = 1,alpha =0.4) 
```
