```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style 
%matplotlib inline
```

```python
random = np.random.rand(10)
```

```python
random
```

```python
style.available
```

```python
style.use('seaborn-whitegrid')
plt.plot(random, 'red' , label = 'line_1' , linewidth = 2 , linestyle  = '--')
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('Plot_1',loc='center')
plt.legend()
plt.show()
```
